### 주어진 iterable 객체의 원소를 이용하여, 길이 n인 조합 가능한 모든 iterable 객체들을 반환해주는 함수를 작성하세요

```python
def subsequences(sub, n):
    
    # check whether n is integer
    if type(n) != type(1): return 'n should be integer'
    
    # iterater through sub with n time
    
    # n == 1 이면 loop 을 돌 필요가 없기 때문에 바로 반환합니다.
    if n == 1 : return [a for a in sub]
    
    # n == len(sub) 이면 그대로 반환
    if n == len(sub) : return tuple([sub])
    
    # 문자열에서 중복을 제거하고 
    sub = sorted(list(sub))
    print(sub)
    
    result = []
    
    def bigger(e, sub):
        return [a for a in sub if e[-1] < a]
    
    def same(e, sub):
        return [a for a in sub if e[-1] == a]
        
    def make_new_sub(x, sub):
        r = []
        for k in bigger(x, sub):
            if x+k not in r:
                r.append(x+k)
        return r
    
    for loop in range(n-1):
        print("-------------")
        print('loop : ', loop+1)
        print("-------------")
        
        if loop == 0:
            for i in range(len(sub)):
                print('sub[i] : ', sub[i])
                print('bigger than sub[i]', bigger(sub[i], sub))
                
                # count sub[i] in string
                print('same as sub[i]', same(sub[i], sub))

                new_sub = make_new_sub(sub[i], sub)
                
                if len(same(sub[i], sub)) > 1:
                    new_sub.append(sub[i] * (loop+2))
                    
                result.extend(new_sub)

                print('new_sub : ', new_sub)
                print('result : ', result)
                print('\n')
        else:
            for i in range(len(result)):
                print('result[i] : ', result[i])
                print('bigger than result[i]', bigger(result[i], sub))

                new_sub = make_new_sub(result[i], sub)
                
                same_num = len(same(sub[i], sub))
                
                if same_num > 1 and result[i].count(sub[i]) < same_num and same_num > (loop+1):
                    new_sub.append(sub[i] * (loop+2))
                    
                result.extend(new_sub)

                print('new_sub : ', new_sub)
                print('result : ', result)
                print('\n')
                
        result = list(set(result))
        print('\n', result)
                
    result = [a for a in result if len(a) == n]

    return tuple(set(result))
```

```python
a = subsequences('BDAABB', 3)
print(a)
```
